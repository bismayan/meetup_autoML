Some basic operations showing inter-operability of H2O Flow and Python/Jupyter notebook

First we import the h2o module and start up an H2o cluster instance

```python
import h2o
```

```python
h2o.init()
```

We can load up a model or a frame used / saved in H2O Flow from Python as well

```python
gbm_top=h2o.get_model("GBM_grid_0_AutoML_20171012_161451_model_1")
```

```python
test_frame=h2o.get_frame("test.hex")
```

```python
valid_frame=h2o.get_frame("train.hex_0.20")
```

Accessing the model performance statistics of the retrieved model on our validation set

```python
perf=gbm_top.model_performance(valid_frame)
```

```python
perf
```

We can get the predictions on a dataframe by using the predict method

```python
pred_test=gbm_top.predict(test_frame)
```

We now combine the predictions of the model with the original dataset by using the cbind method

```python
test_comb=test_frame.cbind(pred_test)
```

```python
test_comb.head()
```

H2O offers the ability to convert H2o Frames to Pandas dataframes via the as_data_frame method.

```python
test_pand=test_comb.as_data_frame()
```

Finally We save the results of the test set in the format required by Kaggle by using Pandas. 

```python
test_pand[["Id","predict"]].rename(columns={"predict":"SalePrice"}).to_csv("autoML1_test_pred.csv",index=False)
```

H2o also allows us to save the models using the save model function

```python
h2o.save_model(gbm_top,path="./automL_first_topGBM")
```

We can access the attributes available in Flow by calling Member functions of the model

```python
gbm_top.varimp_plot(num_of_features=50)
```

In the next few cells we average the predictions of the top 5 models of some pretty long runs of H2o's AutoML to obtain the results that allowed us to place within the top 30% of entries in the Kaggle competition. The steps are the same as above except we average the results of the top 5 models

Note that the models mentioned here are saved in the extended_autoML_models folder so they can be loaded up again. instead of the get_model function, we have to use something like 

h2o.load_model("../extended_autoML_models/GBM_grid_0_AutoML_20171012_161451_model_1")

```python
dl_top=h2o.get_model("DL_grid_0_AutoML_20171012_173310_model_0")
dl_2=h2o.get_model("DL_grid_0_AutoML_20171012_173310_model_4")
gbm_3=h2o.get_model("GBM_grid_1_AutoML_20171012_173310_model_30")
gbm_4=h2o.get_model("GBM_grid_1_AutoML_20171012_182138_model_30")
dl_5=h2o.get_model("DL_grid_0_AutoML_20171012_182138_model_8")
pred_test_dl=dl_top.predict(test_frame)
pred_test_dl2=dl_2.predict(test_frame)
pred_test_gbm3=gbm_3.predict(test_frame)
pred_test_gbm4=gbm_4.predict(test_frame)
pred_test_dl5=dl_5.predict(test_frame)
pred_avge5=(pred_test_dl["predict"]+pred_test_dl2["predict"]+pred_test_gbm3["predict"]+pred_test_gbm4["predict"]\
           +pred_test_dl5["predict"])/5
test_comb_avge5=test_frame.cbind(pred_avge5)
test_pand_avge5=test_comb_avge5.as_data_frame()
test_pand_avge5[["Id","predict"]].rename(columns={"predict":"SalePrice"}).to_csv("autoML5_test_pred.csv",index=False)
```

```python
h2o.save_model(gbm_3,path="./extended_autoML_models")
h2o.save_model(gbm_4,path="./extended_autoML_models")
h2o.save_model(dl_2,path="./extended_autoML_models")
h2o.save_model(dl_5,path="./extended_autoML_models")
```
